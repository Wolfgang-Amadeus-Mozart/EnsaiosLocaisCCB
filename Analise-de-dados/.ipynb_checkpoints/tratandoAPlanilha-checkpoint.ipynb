```python
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta, SU, SA, FR
import calendar
```

```python
df = pd.read_excel('EnsaiosGuarulhos.xlsx')
df
```

```python
#removendo colunas de nome unnamed
for i in range(4,8):
  df.drop(columns=f"Unnamed: {i}", inplace=True)
```

```python
df = df.dropna()
```

```python
#csv apresentava texto que na teoria deveria ser iguais, mas apresentavam espaÃ§os e caracteres especiais que nao deveriam estar ali
df['Semana'] = df['Semana'].str.strip()
df['Semana'] = df['Semana'].str.replace(r'\xa0', ' ', regex=True)
df['Semana'] = df['Semana'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
#checando se a tratativa de texto deu resultado
df['Semana'].unique()
```

```python
def retorna_datas(ano, mes, df):
    datas = []
    # Usando o dicionÃ¡rio de meses, de forma bem rÃºstica mesmo, nÃ£o soube criar de outra forma
    meses_pt = {
        1: 'Janeiro', 2: 'Fevereiro', 3: 'MarÃ§o', 4: 'Abril',
        5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto',
        9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'
    }

    for i in df['Semana']:
        # Aqui eu nao podia perder tempo, na minha cabeÃ§a surgiu uma lÃ³gica de if else e usei, poderia ter usado switch, mas usei aquilo que eu sabia que iria funcionar, atÃ© pq por se tratar de uma analise  rÃ¡pida, nao teria tanto impacto
        if i == 'Primeiro Domingo':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SU(1)))
        elif i == 'Terceiro Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(3)))
        elif i == 'Ãšltimo Domingo':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SU(-1)))
        elif i == 'Primeiro Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(1)))
        elif i == 'Primeira Sexta':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=FR(1)))
        elif i == 'Segundo Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(2)))
        elif i == 'Quarto Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(4)))
        elif i == 'Ãšltimo Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(-1)))
        elif i == 'Terceira Sexta':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=FR(3)))
        elif i == 'Quarto Domingo':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SU(4)))
        elif i == 'Segundo Sexta':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=FR(2)))
        else:
            # Se nÃ£o cair em nenhum, adiciona vazio para nÃ£o quebrar o tamanho da lista
            datas.append(None)

    nome_mes = meses_pt[mes]
    df[f'Data {nome_mes}'] = datas
    return df

ano = 2026
#Todos os meses
for i in range(3,13):
  df = retorna_datas(ano, i, df)

#alterando formato
# Lista todas as colunas que contÃªm a palavra 'Data'
colunas_data = [col for col in df.columns if 'Data' in col]

for col in colunas_data:
    # O .dt.strftime converte para string no formato brasileiro
    df[col] = df[col].dt.strftime('%d/%m/%Y')


df
```
