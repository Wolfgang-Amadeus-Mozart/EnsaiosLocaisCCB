```python
def retorna_datas(ano, mes, df):
    datas = []
    # Usando o dicion치rio de meses, de forma bem r칰stica mesmo, n칚o soube criar de outra forma
    meses_pt = {
        1: 'Janeiro', 2: 'Fevereiro', 3: 'Mar칞o', 4: 'Abril',
        5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto',
        9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'
    }

    for i in df['Semana']:
        # Aqui eu nao podia perder tempo, na minha cabe칞a surgiu uma l칩gica de if else e usei, poderia ter usado switch, mas usei aquilo que eu sabia que iria funcionar, at칠 pq por se tratar de uma analise  r치pida, nao teria tanto impacto
        if i == 'Primeiro Domingo':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SU(1)))
        elif i == 'Terceiro Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(3)))
        elif i == '칔ltimo Domingo':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SU(-1)))
        elif i == 'Primeiro Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(1)))
        elif i == 'Primeira Sexta':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=FR(1)))
        elif i == 'Segundo Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(2)))
        elif i == 'Quarto Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(4)))
        elif i == '칔ltimo Sabado':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SA(-1)))
        elif i == 'Terceira Sexta':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=FR(3)))
        elif i == 'Quarto Domingo':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=SU(4)))
        elif i == 'Segundo Sexta':
            datas.append(datetime(ano, mes, 1) + relativedelta(weekday=FR(2)))
        else:
            # Se n칚o cair em nenhum, adiciona vazio para n칚o quebrar o tamanho da lista
            datas.append(None)

    nome_mes = meses_pt[mes]
    df[f'Data {nome_mes}'] = datas
    return df

ano = 2026
#Todos os meses
for i in range(3,13):
  df = retorna_datas(ano, i, df)

#alterando formato
# Lista todas as colunas que cont칡m a palavra 'Data'
colunas_data = [col for col in df.columns if 'Data' in col]

for col in colunas_data:
    # O .dt.strftime converte para string no formato brasileiro
    df[col] = df[col].dt.strftime('%d/%m/%Y')


df
```
